```python
# Mohammad Sepehr
# A simple project in python about different features affecting a movie's gross revenue.
# dataset: https://www.kaggle.com/datasets/danielgrijalvas/movies
#Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

#Adjusting the configuration of the plots that we'll create.
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#Read the data

df = pd.read_csv('/Users/mohammadsepehr/Desktop/cv/portfolio/movie_correlation/movies.csv')
pd.set_option('display.max_rows', None) #Shows the whole dataset and lifts the limit of 20 rows.
```

```python
#Let's Look at the data
df.head()
```

```python
#Let's see if there is any missing data by looping through the data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull()) 
    print('{}-{}%'.format(col, pct_missing))
```

```python
#For the sake of simplicity, we will drop the missing values.
df = df.dropna()
for col in df.columns:
    pct_missing = np.mean(df[col].isnull()) 
    print('{}-{}%'.format(col, pct_missing))
```

```python
#Let's look at the datatype of each column
df.dtypes
```

```python
# Let's look at each company's Gross Sum Revenue. 
# Hypothesizing that maybe company has a high correlation with Gross.
CompanyGross = df.groupby(['company', 'year'])[["gross"]].sum()

CompanyGrossSorted = CompanyGross.sort_values(['gross','company','year'], ascending = False)[:15]

CompanyGrossSorted = CompanyGrossSorted['gross'].astype('int64') 

CompanyGrossSorted
```

```python
#Dropping the decimal on the budget and gross columns
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
#Sorting and fixing small details to make the data easier to work with.
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(str)
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df.head()
```

```python
#Data is cleaned like we want it. Now we can focus on finding correlations!!!
```

```python
#Scatter plot of budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Revenue')
plt.xlabel('Budget for Movie')
plt.ylabel('Gross Revenue')
plt.show()
```

```python
#Plotting Budget vs Gross using Seaborn

sns.regplot(x='budget', y='gross', data=df, line_kws={'color':'black'})
plt.show()
```

```python
#Let's start looking at correlations!
```

```python
df.corr(method='pearson') #pearson, kendall, spearman
```

```python
#High correlation between budget and gross revenue!
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Let's look at the companies. 
# We need to convert them and have a numeric representation for each company.

df_numeric = df

for col in df_numeric.columns:
    if(df_numeric[col].dtype == 'object'):
        df_numeric[col]= df_numeric[col].astype('category')
        df_numeric[col] = df_numeric[col].cat.codes
        
df_numeric.head()
```

```python
correlation_matrix = df_numeric.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for All Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numeric.corr()
```

```python
correlation_matrix = df_numeric.corr()
corr_pairs = correlation_matrix.unstack()
corr_pairs
```

```python
sorted_corr = corr_pairs.sort_values()
sorted_corr
```

```python
high_corr = sorted_corr[(sorted_corr) > 0.5]
high_corr
```

```python
# Votes and Budget have the highest correlation to the Gross revenue of a movie.
# Company has low correlation with the Gross revenue.
```
